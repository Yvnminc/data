```python
# Read files
import pandas as pd
file_name = "Mobile Device Data for Assignment 2.xlsx"
xl_file = pd.ExcelFile(file_name)

dfs = {sheet_name: xl_file.parse(sheet_name) 
          for sheet_name in xl_file.sheet_names}

dfs.keys()
```

```python
# Assignment
product = dfs['Normalized Product Data']
model = dfs['Model-Company']
company = dfs['Company ID']
```

```python
# Check data
product
```

```python
model = model.dropna()
model
```

```python
company = company.dropna().drop(columns= ["Company"])
company
```

```python
product.isna().sum()
```

```python
model.isna().sum()
```

```python
company.isna().sum()
```

```python
# remove dupicate
company = company.drop_duplicates()
company
```

```python
# Add rows
company.loc[len(company.index)] = ["Other", 0.0]
company
```

```python
company = company.sort_values(by= ["Company_ID"])
company
```

```python
# Joins
product_model = pd.merge(product,model,on='Model',how='left')
```

```python
product_model.isna().sum()
```

```python
# Remove y
drop_list = ["Model ID_y", "Release Date_y", "Release Year_y"]
product_model = product_model.drop(columns=drop_list)
product_model.rename({'Model ID_x': 'Model ID', 'Release Date_x': 'Release Date', 'Release Year_x':'Release Year'}, axis=1, inplace=True)
```

```python
# Fill null
product_model.Company_ID.fillna(0.0,inplace=True)
```

```python
# Show data
product_model
```

```python
# Merge with company
product_model_company = pd.merge(product_model,company,on='Company_ID',how='inner')
product_model_company
```

```python
# Sort by year
product_model_company = product_model_company.sort_values(by = ["Release Year"])
```

```python
product_model_company.isna().sum()
```

```python
product_model_company.to_csv("model.csv", index= False)
```
